```python
load('../common_lib.sage')
load('../NullBasis.sage')

import pandas as pd

ans = {}
for num in range(1,201):
    seed = num
    QUESTION, SOLUTION, CHECKCODE = NullBasis(param=(3,4,2), seed=seed)
    HEADING = r"""\centerline{Quiz 1 \hfill MATH 103 / GEAI 1215: Linear Algebra I}"""
    KEY = r"""NullBasis %3s"""%seed
    
    out_name_q = 'Quiz1-%s.tex'%num
    out_name_s = 'Quiz1-%s_solution.tex'%num

    tex_writer(out_name=out_name_q, sol=False, seed=seed, pdf=True)
    tex_writer(out_name=out_name_s, sol=True, seed=seed, pdf=True)
    
    ans[KEY] = CHECKCODE
    
    s = pd.Series(ans).rename('CHECKCODE')
    s.index.rename('KEY', inplace=True)
    s.to_csv('Quiz1_ans.csv', header=True)
```
